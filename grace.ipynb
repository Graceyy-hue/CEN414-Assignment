```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Measurable Bright Sunshine Source.csv')
df = df.set_index('Country')
del df['S/N']
df
```

```python
for year in range(2010,2020):
    fig,ax = plt.subplots(1)
    df[str(year)].plot.barh(ax=ax, figsize=(10,10), title=f"Number of Days with Measureable Bright Sunshine on Year {year}")
    ax.set_xlabel('Number of Days of Measurable Bright Sunshine')
    
```

```python
vals = []
for year in range(2010,2020):
    vals.append(df[str(year)].mean())
    
plt.plot(range(2010,2020), vals)
plt.title('Mean')
plt.xlabel('Year')
plt.ylabel('Percentage')

```

```python
vals = []
for year in range(2010,2020):
    vals.append(df[str(year)].median())
    

plt.title('Median')
plt.xlabel('Year')
plt.ylabel('Percentage')    
plt.plot(range(2010,2020), vals)
```

```python
vals = []
for year in range(2010,2020):
    vals.append(df[str(year)].std())
    

plt.title('Stdev')
plt.xlabel('Year')
plt.ylabel('Percentage')    
plt.plot(range(2010,2020), vals)
```

```python
vals = []
for year in range(2010,2020):
    vals.append(df[str(year)].var())
    

plt.title('Var')
plt.xlabel('Year')
plt.ylabel('Percentage')    
plt.plot(range(2010,2020), vals)
```

```python
def iqr(sample):
    return sample.quantile(0.75) - sample.quantile(0.25)

vals = []
for year in range(2010,2020):
    vals.append(iqr(df[str(year)]))
    

plt.title('IQR')
plt.xlabel('Year')
plt.ylabel('Percentage')    
plt.plot(range(2010,2020), vals)
```

```python
def siqr(sample):
    return (sample.quantile(0.75) - sample.quantile(0.25))/2

vals = []
for year in range(2010,2020):
    vals.append(siqr(df[str(year)]))
    

plt.title('SIQR')
plt.xlabel('Year')
plt.ylabel('Percentage')    
plt.plot(range(2010,2020), vals)
```

```python
def _range(sample):
    return sample.max() - sample.min()

vals = []
for year in range(2010,2020):
    vals.append(_range(df[str(year)]))
    

plt.title('Range')
plt.xlabel('Year')
plt.ylabel('Percentage')    
plt.plot(range(2010,2020), vals)
```
